WEEK 2 PROJECT-EXPLORATORY DATA ANALYSIS

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. LOAD DATASET

```python
df = pd.read_excel("Dataset for Data Analytics.xlsx")

print("Dataset Loaded Successfully\n")
```

2.BASIC INFORMATION

```python
print("First 5 Rows:")
print(df.head())

```

```python
print("\nDataset Shape:")
print(df.shape)

```

```python
print("\nColumn Information:")
print(df.info())
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

3. CHECK MISSING VALUES

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

 4. CHECK DUPLICATES

```python
duplicates = df.duplicated().sum()
print("\nDuplicate Rows:", duplicates)
```

5. BASIC STATISTICS

```python
print("\nMean of Quantity:")
print(df["Quantity"].mean())
```

```python
print("\nMedian of Quantity:")
print(df["Quantity"].median())
```

```python
print("\nMean of Total Price:")
print(df["TotalPrice"].mean())
```

```python
print("\nMedian of Total Price:")
print(df["TotalPrice"].median())

```

```python
print("\nCount of Orders:")
print(df["OrderID"].count())
```

6. PRODUCT ANALYSIS

```python
print("\nProduct Frequency:")
print(df["Product"].value_counts())
```

```python
# Bar Chart
plt.figure(figsize=(8,5))
df["Product"].value_counts().plot(kind='bar')
plt.title("Product Orders")
plt.xlabel("Product")
plt.ylabel("Number of Orders")
plt.show()
```

7. PAYMENT METHOD ANALYSIS

```python
print("\nPayment Method Distribution:")
print(df["PaymentMethod"].value_counts())
```

```python
plt.figure(figsize=(6,6))
df["PaymentMethod"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Methods")
plt.ylabel("")
plt.show()
```

8. REFERRAL SOURCE ANALYSIS

```python
print("\nReferral Source Distribution:")
print(df["ReferralSource"].value_counts())
```

```python
plt.figure(figsize=(8,5))
df["ReferralSource"].value_counts().plot(kind='bar')
plt.title("Referral Sources")
plt.xlabel("Source")
plt.ylabel("Customers")
plt.show()
```

9. MONTHLY SALES TREND

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Month"] = df["Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["TotalPrice"].sum()

print("\nMonthly Sales:")
print(monthly_sales)
```

```python
plt.figure(figsize=(12,5))
monthly_sales.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

10. TOTAL PRICE DISTRIBUTION

```python
plt.figure(figsize=(8,5))
plt.hist(df["TotalPrice"], bins=20)
plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.show()
```

 11. OUTLIER DETECTION

```python
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df["TotalPrice"] < lower_bound) |
    (df["TotalPrice"] > upper_bound)
]

print("\nNumber of Outliers:")
print(len(outliers))

print("\nOutlier Records:")
print(outliers)
```

```python
# Box Plot
plt.figure(figsize=(8,5))
plt.boxplot(df["TotalPrice"])
plt.title("Box Plot of Total Price")
plt.show()
```

12. ORDER STATUS ANALYSIS

```python
print("\nOrder Status:")
print(df["OrderStatus"].value_counts())

plt.figure(figsize=(7,5))
df["OrderStatus"].value_counts().plot(kind='bar')
plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
top_orders = df.nlargest(10, "TotalPrice")

print("\nTop 10 High Value Orders:")
print(top_orders[["OrderID", "Product", "TotalPrice"]])
```
